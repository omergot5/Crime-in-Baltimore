```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# קישורי הקבצים הגולמיים (raw) מ-GitHub
url1 = 'https://raw.githubusercontent.com/omergot5/Crime-in-Baltimore/f1f70aa586f620cc6860651bcd4ccea395aeb27b/Bol1.csv'
url2 = 'https://raw.githubusercontent.com/omergot5/Crime-in-Baltimore/f1f70aa586f620cc6860651bcd4ccea395aeb27b/Bol2.csv'
url3 = 'https://raw.githubusercontent.com/omergot5/Crime-in-Baltimore/f1f70aa586f620cc6860651bcd4ccea395aeb27b/Bol3.csv'
url4 = 'https://raw.githubusercontent.com/omergot5/Crime-in-Baltimore/f1f70aa586f620cc6860651bcd4ccea395aeb27b/Bol4.csv'

# קריאת הקבצים לקבצי DataFrame
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
df4 = pd.read_csv(url4)

# איחוד הקבצים על ידי הצמדת שורות
combined_df = pd.concat([df1, df2, df3, df4], ignore_index=True)

# שמירת הקובץ המשולב לקובץ CSV חדש
combined_df.to_csv('combined_file.csv', index=False)


combined_df.head()
```
